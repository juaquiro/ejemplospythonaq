```python
import numpy as np
import cv2
import os
from matplotlib import pyplot as plt
#built-in magic command Set up matplotlib to work interactively.
%matplotlib notebook
```

```python
NR, NC, D=(128, 127, 1)
```

```python
x=np.arange(0, NC)
x=x-0.5*NC
y=np.arange(0, NR)
y=y-0.5*NR
```

```python
X,Y=np.meshgrid(x,y)
```

```python
Z=abs(X+1j*Y)
```

```python
h1 = plt.contourf(X,Y,Z, cmap=plt.get_cmap('hot'))
plt.show()
```

```python
h2=plt.imshow(Z, cmap=plt.get_cmap('hot'))
plt.show()
```

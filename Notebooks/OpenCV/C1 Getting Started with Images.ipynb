## nota 1.  sobre cv2.imshow y jupyter
cuidado, si no se cierra la vantana apropiadamente, el kernel se cuelga. Como se explica en https://stackoverflow.com/questions/46236180/opencv-imshow-will-cause-jupyter-notebook-crash el motivo es que "cv.imshow doesn't really make sense in a client/server environment like Jupyter. You want something that will display the image as part of the notebook (on the client side), not to run a GUI window on the server side -- that might kinda "work" when the client and server are on the same machine, but that's about it. | You already use the imshow function from matplotlib (not numpy as you seem to think)"

```python
#load image in original format and display it, hacer click en la ventana y presionar cualquier tecla
img = cv2.imread(fileName,cv2.IMREAD_COLOR)
cv2.namedWindow('image', cv2.WINDOW_AUTOSIZE)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## nota 2. Color image loaded by OpenCV is in BGR mode. But Matplotlib displays in RGB mode

```python
b,g,r = cv2.split(img)
img2 = cv2.merge([r,g,b])
plt.subplot(121);plt.imshow(img) # expects distorted color
plt.subplot(122);plt.imshow(img2) # expect true color
plt.show()
```

```python
#de forma alternativa jugamos con el indexado de numpy
img2 = img[:,:,::-1]
plt.subplot(121);plt.imshow(img) # expects distorted color
plt.subplot(122);plt.imshow(img2) # expect true color
plt.show()
```

```python
# Load an color image in grayscale, rescale it and display
img = cv2.imread(fileName,cv2.IMREAD_GRAYSCALE)
dim1 = img.shape
scale=6.3
# transform tuple to np array, dot multiply cast to int array and back to tuple
dim2=tuple((np.rint(scale*np.asarray(dim1))).astype(int))
```

```python
type(dim1)
```

```python
type(dim2)
```

```python
img, dim1, dim2
```

```python
#resize image
rimg = cv2.resize(img,dim2)
```

```python
#create a figure and display the image in it
cv2.namedWindow('image', cv2.WINDOW_AUTOSIZE)
cv2.imshow('image',rimg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## nota 2. para poder mostrar la imagen "en linea" dentro del notebook hay que usar imwhow de matplotlib

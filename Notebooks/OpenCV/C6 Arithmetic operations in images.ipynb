# Arithmetic Operations on Images
de https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_core/py_image_arithmetics/py_image_arithmetics.html

```python
#init modules
import numpy as np
import cv2
import os
from matplotlib import pyplot as plt

#set basic vars
baseDir=os.path.join('..','..','ExamplesRepo')
#file name
baseFileNameList=('cameraman.tif', 'eight.tif', 'circuit.tif')
```

```python
# load images using a list

(cameraman, eight, circuit)=(0,1,2)
n=0
#a list of 3x1 0s
imgList=[0]*3
for name in baseFileNameList:
    fileName=os.path.join(baseDir, name)
    imgList[n] = cv2.imread(fileName,cv2.IMREAD_GRAYSCALE)    
    n+=1
for im in imgList:
    plt.figure()
    plt.imshow(im, cmap=plt.cm.gray)
    plt.show()
```

```python
#the same than before but usong a dictionary
#using dict.fromkeys
#imgDict=dict.fromkeys(baseFileNameList, None)
#using a comprehension solution 
imgDict={name: None for name in baseFileNameList}
```

```python
for name in baseFileNameList:
    fileName=os.path.join(baseDir, name)
    imgDict[name] = cv2.imread(fileName,cv2.IMREAD_GRAYSCALE)    
    
for name in baseFileNameList:    
    plt.figure()
    im=imgDict[name]
    plt.imshow(im, cmap=plt.cm.gray)
    plt.show()
```

## Image Addition
cuidado openCV es saturado y numpy es modulo

```python
#numpy is modulo
cam=imgList[cameraman]
cam2=100+cam
plt.imshow(cam2, cmap=plt.cm.gray)
```

```python
#openCV is saturated
cam2=cv2.add(100,cam)
plt.imshow(cam2, cmap=plt.cm.gray)
```

## Resize operation
lets resize all images to the same size

```python
NR, NC=641, 451
imgDictResize=imgDict
for name in baseFileNameList:    
    im=imgDict[name]
    imgDictResize[name]=cv2.resize(im, (NR, NC), interpolation = cv2.INTER_LINEAR)
    
```

## Image Blending

```python
#define weights
alpha, beta, gamma=0.4, 0.5, 0.6
```

```python
imBlend=cv2.addWeighted(imgDict['cameraman.tif'], alpha, imgDict['eight.tif'], beta, gamma)
plt.figure()
plt.imshow(imBlend, cmap=plt.cm.gray)
plt.show()
```
